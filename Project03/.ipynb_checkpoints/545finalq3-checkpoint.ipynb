```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import norm, skewnorm, norminvgauss
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

ticker = "SPY"
data = yf.download(ticker, start="2010-01-01", end="2023-01-01", progress=False)


if 'Adj Close' in data.columns:
    price_column = 'Adj Close'
elif 'Close' in data.columns:
    price_column = 'Close'
else:
    raise KeyError("Neither 'Adj Close' nor 'Close' columns are available in the data.")

data['LogReturn'] = np.log(data[price_column] / data[price_column].shift(1))
returns = data['LogReturn'].dropna()
mu, sigma = norm.fit(returns)
a_skew, loc_skew, scale_skew = skewnorm.fit(returns)
params_nig = norminvgauss.fit(returns)
a_nig, b_nig, loc_nig, scale_nig = params_nig
def compute_loglikelihood(pdf_vals):
    return np.sum(np.log(pdf_vals))

def compute_AIC(logL, k):
    return 2 * k - 2 * logL

pdf_normal = norm.pdf(returns, loc=mu, scale=sigma)
pdf_skew = skewnorm.pdf(returns, a_skew, loc=loc_skew, scale=scale_skew)
pdf_nig = norminvgauss.pdf(returns, a_nig, b_nig, loc=loc_nig, scale=scale_nig)

logL_normal = compute_loglikelihood(pdf_normal)
logL_skew = compute_loglikelihood(pdf_skew)
logL_nig = compute_loglikelihood(pdf_nig)

AIC_normal = compute_AIC(logL_normal, k=2)
AIC_skew = compute_AIC(logL_skew, k=3)
AIC_nig = compute_AIC(logL_nig, k=4)

print("Fitted Parameters:")
print("Normal:\t\t mu = {:.5f}, sigma = {:.5f}".format(mu, sigma))
print("Skew Normal:\t a = {:.5f}, loc = {:.5f}, scale = {:.5f}".format(a_skew, loc_skew, scale_skew))
print("NIG:\t\t a = {:.5f}, b = {:.5f}, loc = {:.5f}, scale = {:.5f}".format(a_nig, b_nig, loc_nig, scale_nig))

print("\nAIC Comparison:")
print("Normal AIC:\t\t {:.2f}".format(AIC_normal))
print("Skew Normal AIC:\t {:.2f}".format(AIC_skew))
print("NIG AIC:\t\t {:.2f}".format(AIC_nig))

plt.figure(figsize=(10, 6))
plt.hist(returns, bins=50, density=True, alpha=0.5, label='Empirical Data')

x = np.linspace(returns.min(), returns.max(), 1000)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), label='Normal PDF', linewidth=2)
plt.plot(x, skewnorm.pdf(x, a_skew, loc=loc_skew, scale=scale_skew), label='Skew-Normal PDF', linewidth=2)
plt.plot(x, norminvgauss.pdf(x, a_nig, b_nig, loc=loc_nig, scale=scale_nig), label='NIG PDF', linewidth=2)

plt.title("Fitted Distributions to Daily Log Returns of " + ticker)
plt.xlabel("Daily Log Return")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
